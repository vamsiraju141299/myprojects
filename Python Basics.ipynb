2. Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known
as twin primes


```
def twinprime(n): 
      
    k = [True for i in range(n + 2)] 
    p = 2
      
    while (p * p <= n + 1): 
          
        if (k[p] == True): 
              
            for i in range(p * 2, n + 2, p): 
                k[i] = False
        p += 1
       
    for p in range(2, n-1): 
        if k[p] and k[p + 2]: 
            print("(",p,",", (p + 2), ")" ,end='') 
  
  
if __name__=='__main__': 
  n=int(input())
  twinprime(n)
 
```

7.Write a function prodDigits() that inputs a number and returns the product of digits of that number.

```
def prodDigits(n): 
  
    product = 1
  
    while (n != 0): 
        product = product * (n % 10) 
        n = n // 10
  
    return product 
  
n = int(input())
print(prodDigits(n)) 
  
```

9.Write a function sumPdivisors() that finds the sum of proper divisors of a number. Proper divisors of a number are those
numbers by which the number is divisible, except the number itself. For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 18

```
import math
def sumPdivisors(n):
  l=[]
  for i in range(1,int(math.sqrt(n))+1,1):
    if(n%i==0):
      l.append(i)
      if(i!=(n//i)):
        l.append(n//i)
  l.sort()
  l.pop()
  for x in range(len(l)-1):
    print(l[x],end=',')
  print(l[len(l)-1])
  print("sum of the proper divisors is",sum(l))
sumPdivisors(36)

```

10.A number is called perfect if the sum of proper divisors of that number is equal to the number. For example 28 is perfect
number, since 1+2+4+7+14=28. Write a program to print all the perfect numbers in a given range

```
 
def isPerfect( n ): 
    sum = 1
    i = 2
    while i * i <= n: 
        if n % i == 0: 
            sum = sum + i + n/i 
        i += 1
      
    return (True if sum == n and n!=1 else False) 
n = int(input())
k=int(input())
for n in range (k): 
    if isPerfect (n): 
        print(n, end=',')
```

11.Two different numbers are called amicable numbers if the sum of the proper divisors of each is equal to the other number.
For example 220 and 284 are amicable numbers. Sum of proper divisors of 220 = 1+2+4+5+10+11+20+22+44+55+110 = 284
Sum of proper divisors of 284 = 1+2+4+71+142 = 220 Write a function to print pairs of amicable numbers in a range


```
def sumPdivisors(n):
  l=[]
  for i in range(1,int(math.sqrt(n))+1,1):
    if(n%i==0):
      l.append(i)
      if(i!=(n//i)):
        l.append(n//i)
  l.sort()
  l.pop()
  return sum(l)
l=[]
for i in range(200,221,1):
  r=sumPdivisors(i)
  p=sumPdivisors(r)
  if(i==p):
    l.append((r,p))
print(l)
for i in l:
  print("pair of amicable numbers",i)
```
